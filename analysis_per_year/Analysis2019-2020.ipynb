# MSc & BSc

```python
xls = pd.ExcelFile('DataAnalysis/All_MSc_BSc.xlsx')
sheets = pd.read_excel(xls, sheet_name=None)
frames = list(sheets.values())[1:]
df1 = pd.concat(frames)
```

```python
profile1 = df1.profile_report(title='MSc & BSc Report', check_correlation_pearson=False)
profile1.to_file(output_file="reports/MScBSc.html")
```
